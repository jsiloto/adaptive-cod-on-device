```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("results_dynamic_bitrate_s4.csv", header=None, delimiter=" ")
# print(df)
print(df.iloc[:, 0])
configurable = np.array([ [1.0, 16.6], [2.0, 29.7], [3.0, 34.6], [4.0, 36.8]]).T


plt.plot(df.iloc[:, 1], df.iloc[:, 0]*100, label="Continuous", marker='.', color='b', zorder=3)
plt.scatter(configurable[0], configurable[1], label="Discrete", color='black', zorder=3)
plt.legend()
plt.grid(zorder=0)
plt.savefig("flexrate.pdf")
```
